```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
data=pd.read_csv('fuel-econ.csv')
data
```

```python
data.info()
```

```python
sns.set(style="whitegrid")
```

```python
# 1. Distribution of Fuel Economy (Combined MPG)
plt.figure(figsize=(10, 6))
sns.histplot(data['comb'], bins=30, kde=True, color='skyblue')
plt.title('Distribution of Combined Fuel Economy (MPG)')
plt.xlabel('Combined MPG')
plt.ylabel('Frequency')
plt.show()
```

```python
# 2. Fuel Economy by Vehicle Class
plt.figure(figsize=(14, 8))
sns.boxplot(x='comb', y='VClass', data=data, palette='Set2')
plt.title('Fuel Economy by Vehicle Class')
plt.xlabel('Combined MPG')
plt.ylabel('Vehicle Class')
plt.show()
```

```python
# 3. CO2 Emissions vs. Engine Displacement
plt.figure(figsize=(10, 6))
sns.scatterplot(x='displ', y='co2', hue='fuelType', data=data, palette='muted')
plt.title('CO2 Emissions vs. Engine Displacement')
plt.xlabel('Engine Displacement (L)')
plt.ylabel('CO2 Emissions (g/km)')
plt.show()
```

```python
# 4. Fuel Type Distribution
plt.figure(figsize=(8, 6))
sns.countplot(y='fuelType', data=data, palette='coolwarm')
plt.title('Fuel Type Distribution')
plt.xlabel('Count')
plt.ylabel('Fuel Type')
plt.show()
```

```python
# 4. Fuel Economy Score Distribution
plt.figure(figsize=(10, 6))
sns.histplot(data['feScore'], bins=10, kde=True, color='coral')
plt.title('Fuel Economy Score Distribution')
plt.xlabel('Fuel Economy Score')
plt.ylabel('Frequency')
plt.show()
```

```python
# 5. Fuel Economy vs. CO2 Emissions by Drive Type
plt.figure(figsize=(10, 6))
sns.scatterplot(x='comb', y='co2', hue='drive', data=data, palette='coolwarm')
plt.title('Fuel Economy vs. CO2 Emissions by Drive Type')
plt.xlabel('Combined MPG')
plt.ylabel('CO2 Emissions (g/km)')
plt.show()
```
